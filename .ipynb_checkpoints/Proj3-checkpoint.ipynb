```python
import numpy as np
from surprise import Dataset
from surprise import Reader
from surprise.model_selection.split import train_test_split
import matplotlib.pyplot as plt
from surprise.prediction_algorithms.knns import KNNWithMeans
```

```python
reader = Reader(sep=',')
data = Dataset.load_from_file('./ml-latest-small/ratings_new.csv', reader=reader)
```

```python
data.split(n_folds=10)
sim_options = {'name': 'pearson',\
              'user_based': True\
              }
algo = KNNWithMeans(k=48, sim_options=sim_options)
```

```python
t_all = range(1, 26)
pre_all = np.zeros(25)
rec_all = np.zeros(25)
for trainset, testset in data.folds():
    algo.fit(trainset)
    pred = algo.test(testset)
    G_all = dict()
    S_all = dict()
    for elem in pred:
        if elem.r_ui >= 3:
            G_all.setdefault(elem.uid, set()).add(elem.iid)
        S_all.setdefault(elem.uid, list()).append((elem.est, elem.iid))
    pre = [[] for _ in range(25)]
    rec = [[] for _ in range(25)]
    
    for uid, G in G_all.items():
        S = sorted(S_all[uid], key = lambda x: x[0], reverse = True)
        inter = 0
        for t in t_all:
            if len(S) < t:
                break
            if S[t - 1][1] in G:
                inter += 1
            pre[t - 1].append(inter / t)
            rec[t - 1].append(inter / len(G))
            
    for i in range(25):
        pre_all[i] += np.mean(pre[i])
        rec_all[i] += np.mean(rec[i])

pre_all /= 10
rec_all /= 10
plt.plot(t_all, pre_all)
plt.xlabel('t')
plt.ylabel('precision')
plt.title('precision vs t for KNN')
plt.show()
plt.plot(t_all, rec_all)
plt.xlabel('t')
plt.ylabel('recall')
plt.title('recall vs t for KNN')
plt.show()
plt.plot(pre_all, rec_all)
plt.xlabel('recall')
plt.ylabel('precision')
plt.title('precision vs recall for KNN')
plt.show()
```
